```python
from rag_functions import (
    connect_to_collection,
    query_collection
)

from sentence_transformers import SentenceTransformer
embedding_model_name = "all-mpnet-base-v2"
embedding_model = SentenceTransformer(model_name_or_path=embedding_model_name, 
                                      device="cuda") 
chroma_client, collection = connect_to_collection()

queries = [
    'What can you tell me about the CrowdStrike outage?',
    'What happened at Chicago Ohare regarding Crowdstrike?',
    'I heard there was something that caused a delay at Ohare airport recently. Can you tell me anything about it?'
]



for query in queries:
    context_items, results = query_collection(query, collection, embedding_model,NLP = True)
    print(results.keys())
    print('###############################')
    print('Query:\n ')
    print('###############################')
    print(query)
    print('\t---embedding search---')
    for score, result in zip(results['distances'][0],results['documents'][0]):
        print('\t\tEmbed Distance:',score)
        print()
        print('\t\tSearch result:',result)
        print()
    print('\n---\n')
    print('\t---NLP---')
    for context in context_items:
        print('\t\tScore:',context['score'])
        print()
        print('\t\tDocument:',context['document'])
        print()
    
    
```
